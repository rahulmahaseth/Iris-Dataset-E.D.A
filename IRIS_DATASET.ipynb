COMMULATIVE DATA ANALYSIS:CDF

## counts is the value getting corresponding to histogram value height

```python
#effect of bin value
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density = True)
pdf = counts/(sum(counts))
print("pdf",pdf);
print("bin",bin_edges);
cdf = np.cumsum(pdf)
pyp.plot(bin_edges[1:],pdf);
pyp.plot(bin_edges[1:], cdf)
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=20, density = True)
pdf = counts/(sum(counts))
pyp.plot(bin_edges[1:],pdf);
pyp.show();

```

```python
#print(iris_setosa.petal_length)
print("maximum:",iris_setosa.petal_length.max())
print("minimum:",iris_setosa.petal_length.min())
COUNTS,BIN_EDGES=np.histogram(iris_setosa.petal_length,bins=10,density=True)#return the height value of each histogram and the bin_edges
print("counts:",COUNTS)
pdf=COUNTS/(sum(COUNTS))#probability density function
print("PDF:",pdf);
print("BIN_EDGES:",BIN_EDGES)
#compute CDF
cdf=np.cumsum(pdf)#commutative sum
pyp.plot(BIN_EDGES[1:],pdf) 
pyp.plot(BIN_EDGES[1:],cdf)
pyp.show()
```

```python
## Plots of CDF of petal_length for various types of flowers.

# Misclassification error if you use petal_length only.
#PLOTTING CDF OF PETAL_LENGTH
#int(iris_setosa.petal_length)
print("maximum:",iris_setosa.petal_length.max())
print(iris_setosa.petal_length.min())
COUNTS,BIN_EDGES=np.histogram(iris_setosa.petal_length,bins=10,density=True)
print(COUNTS)
pdf=COUNTS/(sum(COUNTS))
print(pdf);
print(BIN_EDGES)
#compute CDF
cdf=np.cumsum(pdf)
pyp.plot(BIN_EDGES[1:],pdf)
pyp.plot(BIN_EDGES[1:],cdf)
COUNTS,BIN_EDGES=np.histogram(iris_virginica.petal_length,bins=10,density=True)
print(COUNTS)
pdf=COUNTS/(sum(COUNTS))
print(pdf);
print(BIN_EDGES)
#compute CDF
cdf=np.cumsum(pdf)
pyp.plot(BIN_EDGES[1:],pdf)
pyp.plot(BIN_EDGES[1:],cdf)
COUNTS,BIN_EDGES=np.histogram(iris_versicolor.petal_length,bins=10,density=True)
print(COUNTS)
pdf=COUNTS/(sum(COUNTS))
print(pdf);
print(BIN_EDGES)
#compute CDF
cdf=np.cumsum(pdf)
pyp.plot(BIN_EDGES[1:],pdf)
pyp.plot(BIN_EDGES[1:],cdf)
pyp.show
#SAYS ABOUT ERROR IN PREDICITON
#SAYS THE SUCCESS RATE AT ANY POINT
```

MEAN VARIANCE STANDARD DAVIATION

```python
# mean is average value of all data point 
#varince is the average of square of distance of all point from mean value
#std-var is square root of variance
print("means:")
print(np.mean(iris_setosa.petal_length))
#mean with an outlier
print(np.mean(np.append(iris_setosa["petal_length"],50)))
print(np.mean(iris_versicolor.petal_length))
print(np.mean(iris_virginica.petal_length))
print("\nstd-dev");
print(np.std(iris_setosa.petal_length))
print(np.std(iris_versicolor.petal_length))
print(np.std(iris_virginica.petal_length))
```

MEDIAN 

```python
#array must be sorted
#median gives the incorrect data when its more than 50 percent data is correpted
print("MEDIAN:")
print(np.median(iris_setosa.petal_length))
#mean with an outlier
print(np.median(np.append(iris_setosa["petal_length"],50)))
print(np.median(iris_versicolor.petal_length))
print(np.median(iris_virginica.petal_length))
```

percentile and quantile

```python
#array must be sorted
print("quantile")
print(np.percentile(iris_setosa["petal_length"],np.arange(0,100,25)))
print(np.percentile(iris_versicolor.petal_length,np.arange(0,100,25)))
print(np.percentile(iris_virginica.petal_length,np.arange(0,100,25)))
```

```python
#array must be sorted
print("\n90th percentile")
print(np.percentile(iris_setosa["petal_length"],90))
print(np.percentile(iris_versicolor.petal_length,90))
print(np.percentile(iris_virginica.petal_length,90))
```

MEDIAN ABSOLUTE DEVIATION:

```python
# MAD IS CALCULATING HOW FAR IS THE ALL POINT FROM THE MEDIAN AND CALCULATING THE MEDIAN OF THE ABSOLUTE DIFFERENCE
from statsmodels import robust
print("\nMedian standard deviation")
print(robust.mad(iris_setosa.petal_length))
print(robust.mad(iris_virginica.petal_length))
print(robust.mad(iris_versicolor.petal_length))
```
