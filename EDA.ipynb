```python
# ADDING VARIABLE TO THE COLUMN 
data_positive=data_set.loc[data_set["status"]==1]
data_positive
```

```python
data_negative=data_set.loc[data_set.status==2]
data_negative
```

```python
#PLOTTING 1-D SCATTER PLOT
plt.plot(data_positive["nodes"],np.zeros_like(data_positive["nodes"]),'go')
plt.plot(data_negative["nodes"],np.zeros_like(data_negative["nodes"]),'r^')
plt.show()
```

PDF ,CDF

```python
#ANALYSING DISTRIBUTIVE PLOT ON THE BASIS OF PDF(SMMOTHNING),CDF
count,bin_edges=np.histogram(data_positive["nodes"],10,density=True)
print("count",count)
print(bin_edges)
pdf=count/(sum(count))
print("pdf:",pdf)
data_positive.nodes.min()
cdf=np.cumsum(pdf)
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
count,bin_edges=np.histogram(data_negative["nodes"],10,density=True)
print("count",count)
print(bin_edges)
pdf=count/(sum(count))
print("pdf:",pdf)
data_positive.nodes.min()
cdf=np.cumsum(pdf)
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
